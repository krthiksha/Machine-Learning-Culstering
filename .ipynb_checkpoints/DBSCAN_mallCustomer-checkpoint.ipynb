# Step 1 

## read the dataset 

```python
import pandas as pd
dataset = pd.read_csv("Mall_Customers.csv")
dataset
```

## Input 
### Select the 2 important columns (which has the greater influence over the clustering for type of customers)

```python
x = dataset.iloc[:, [3,4]].values
print(x , type(x))
```

## Standardization

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
x
```

## KEY PARAMETERS to be find before model creation

```python
# find the min_samples 
""" 
min_samples = 2 * number of features
min_samples= 2*2 =4 
"""
MinSamples = 4

# find epsilon value using K distance Graph 
from sklearn.neighbors import NearestNeighbors
import numpy as np
import matplotlib.pyplot as plt 

def plot_k_distance_graph(X,k):
    neigh = NearestNeighbors(n_neighbors=k)
    neigh.fit(X)
    distances, indices = neigh.kneighbors(X)
    distances = np.sort(distances[:,k-1])
    plt.figure(figsize=(10,6))
    plt.plot(distances)
    plt.xlabel('Points')
    plt.ylabel(f'{k}-th nearest neighbor distance')
    plt.title('k-distance Graph')
    plt.show()

plot_k_distance_graph(X=x, k=MinSamples)


"""
k-distance Graph -> EPSILON = 0.5
"""
```

## Model Creation 

```python
from sklearn.cluster import DBSCAN 
DBSCAN = DBSCAN(eps=0.5, min_samples=MinSamples)
y_label = DBSCAN.fit_predict(x)
y_label
```

```python
# to list the methods and attributes of any algorithm - use dir() function
dir(DBSCAN)
```

## Create the new dataset by adding  the cluster group (type of customer category)

```python
supervised = dataset 
supervised["cluster_group"] = y_label
supervised  # new dataset 
```

## save the dataset in csv file 

```python
supervised.to_csv("dbscan_cluster.csv",index=False)
```

# Step 3 - Evaluation Metric (clustering)

```python
from sklearn.metrics import davies_bouldin_score
dbi = davies_bouldin_score(x, y_label)
print("Davies-Bouldin Index:", dbi)

""" 
Davies-Bouldin Index:-
0 (best) â†’ Clusters are compact and well-separated.
Closer to 0 â†’ Better clustering performance.
Higher values (>1, 2, etc.) â†’ Poor clustering, clusters overlap or are not distinct.
"""
```
