# BIRCH clustering - Unsupervised learning
## Problem statement: Malls customer dataset (Target Marketting)
https://scikit-learn.org/stable/modules/generated/sklearn.cluster.Birch.html#sklearn.cluster.Birch

# Step 1 

## read the dataset 

```python
import pandas as pd
dataset = pd.read_csv("Mall_Customers.csv")
dataset
```

## Input 
### Select the 2 important columns (which has the greater influence over the clustering for type of customers)

```python
x = dataset.iloc[:, [3,4]].values
print(x , type(x))
```

## Standardization

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
x
```

## Model Creation 

```python
# Model creation
from sklearn.cluster import Birch
Birch_model = Birch(threshold=0.4, branching_factor=50, n_clusters=5)
y_label = Birch_model.fit_predict(x)
print(y_label)

"""
tuning:-
threshold=0.5, branching_factor=50, n_clusters=5   -- cluster 1 was not well separated
threshold=0.4, branching_factor=50, n_clusters=5   -- all clusters are well separated
"""
```

```python
# RESULTS
cf_subcluster_center = Birch_model.subcluster_centers_
print("cf_subcluster_center: ", cf_subcluster_center)
cf_subcluster_labels = Birch_model.subcluster_labels_
print("cf_subcluster_labels: ", cf_subcluster_labels)
```

```python
# to list the methods and attributes of any algorithm - use dir() function
# dir(Birch_model)
```

## Create the new dataset by adding  the cluster group (type of customer category)

```python
supervised = dataset 
supervised["cluster_group"] = y_label
supervised  # new dataset 
```

## save the dataset in csv file 

```python
supervised.to_csv("Birch_cluster.csv",index=False)
```

# Step 3 - Evaluation metric (clustering)

```python
from sklearn.metrics import silhouette_score
score = silhouette_score(x, y_label)
print("Silhouette Score:", score)
# -1 → 1 (higer is better (1))
```
