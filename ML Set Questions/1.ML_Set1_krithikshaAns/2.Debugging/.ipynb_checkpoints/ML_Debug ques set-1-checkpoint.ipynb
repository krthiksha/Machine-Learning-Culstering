```python
#1.Debug the given code
import pandas as pd

data = {'Feature1': ['10', '20', 'Thirty'],  # 'Thirty' is not a valid number
        'Feature2': [5.5, 6.7, 8.9]}

# df['Feature1'] = df['Feature1'].astype(int)  # Error: Cannot convert 'Thirty' to int


""" converting categorical data into numerical using get_dummies """
df = pd.DataFrame(data)
df = pd.get_dummies(df, drop_first=True, dtype=int)
print(df)

```

```python
#2.
import pandas as pd

data = {'A': [1, 2, None],  # Missing value
        'B': [4, None, 6]}  # Missing value

df = pd.DataFrame(data)
""" calculates Mean """
mean_value = df.mean() 
"""
fillna - default creates new dataframe
inplace=True - displays with modification made in dataframe"""
df.fillna(mean_value, inplace=True)  # Error: fillna() does not modify in place

print(df)
```

```python
#7.from sklearn.linear_model import LogisticRegression
import pandas as pd
from sklearn.linear_model import LogisticRegression
X_train = [[1, 2], [3, 4], [5, 6]]
y_train = ["yes", "no", "yes"]  # Error: LogisticRegression expects numerical labels
""" converting categorical into numerical """
y_train = pd.get_dummies(y_train, drop_first=True, dtype=int)

model = LogisticRegression()
model.fit(X_train, y_train)

model.predict([[3,4]])
```

```python
#8.
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# Creating a dataset with missing values
X_train = np.array([[1, 2], [3, np.nan], [5, 6]])
y_train = np.array([10, 20, 30])

""" Replace nan with mean of the column using simple imputer """
from sklearn.impute import SimpleImputer
impute = SimpleImputer(strategy='mean')
x_train = impute.fit_transform(X_train)

model = LinearRegression()
model.fit(x_train, y_train)  

model.predict([[1,2]])
```

```python
#9.
from sklearn.tree import DecisionTreeClassifier

X_train = [[1, 2], [3, 4], [5, 6]]
y_train = ["spam", "ham", "spam"]  # Error: String labels not allowed

""" convert categorical into numerical """
y_train = pd.get_dummies(y_train, drop_first=True, dtype=int)

model = DecisionTreeClassifier()
model.fit(X_train, y_train)


model.predict([[3,4]])
```

```python
#10.
from sklearn.svm import SVC

X_train = [[1, 100], [2, 200], [3, 300]]
y_train = [0, 1, 0]

""" to handle the unreliable output - do the standardization in order to normalize the large scale difference """
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
# print(X_train)

model = SVC()
model.fit(X_train, y_train)
standardize_input = sc.transform([[3, 300]])
print(model.predict(standardize_input))  # Unreliable output due to large-scale difference
```
