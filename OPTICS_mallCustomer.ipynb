## Input

```python
x = dataset.iloc[:, [3,4]].values
print(x , type(x))
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
x
```

# Step 3 - Evaluation metric (clustering)

```python
from sklearn.metrics import davies_bouldin_score
dbi = davies_bouldin_score(x, y_label)
print("Davies-Bouldin Index:", dbi)

""" 
Davies-Bouldin Index:
0 (best) → Clusters are compact and well-separated.
Closer to 0 → Better clustering performance.
Higher values (>1, 2, etc.) → Poor clustering, clusters overlap or are not distinct.
"""
```
