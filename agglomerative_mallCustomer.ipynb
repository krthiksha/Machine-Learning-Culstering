# Agglomerative hierarchical clustering - Unsupervised learning
## Problem statement: Malls customer dataset (Target Marketting)

# Step 1 

## read the dataset 

```python
import pandas as pd
dataset = pd.read_csv("Mall_Customers.csv")
dataset
```

## Input 
### Select the 2 important columns (which has the greater influence over the clustering for type of customers)

```python
x = dataset.iloc[:, [3,4]].values
print(x , type(x))
```

## Finding the No of clusters  - using dendrogram (in agglomerative hierarchical clustering algorithm)

```python
import scipy.cluster.hierarchy as sch 
import matplotlib.pyplot as plt 

dendrogram = sch.dendrogram(sch.linkage(x,method='ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distance')  # this will be calculated by dendrogram
plt.show()

""" 5 big wards in dendrogram  = No of Clusters = 5 (Analyzed from the graph ) """
```

## Model Creation 

```python
from sklearn.cluster import AgglomerativeClustering 
agglo_model = AgglomerativeClustering(n_clusters=5)
y_label = agglo_model.fit_predict(x)
y_label
```

```python
# to list the methods and attributes of any algorithm - use dir() function
dir(agglo_model)
```

## Create the new dataset by adding  the cluster group (type of customer category)

```python
supervised = dataset 
supervised["cluster_group"] = y_label
supervised  # new dataset 
```

## save the dataset in csv file 

```python
supervised.to_csv("agglomerative_cluster.csv",index=False)
```
