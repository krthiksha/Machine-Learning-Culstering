# Step 1 

## read the dataset 

```python
import pandas as pd
dataset = pd.read_csv("Mall_Customers.csv")
dataset
```

## Input 
### Select the 2 important columns (which has the greater influence over the clustering for type of customers)

```python
x = dataset.iloc[:, [3,4]].values
print(x , type(x))
```

## Standardization

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
x
```

## Model Creation 

```python
from sklearn.cluster import HDBSCAN 
hdbscan = HDBSCAN(min_cluster_size=8, min_samples=None )   # tune 5-10 => min_cluster_size
y_label = hdbscan.fit_predict(x)
y_label
```

```python
# to list the methods and attributes of any algorithm - use dir() function
dir(hdbscan)
```

## Create the new dataset by adding  the cluster group (type of customer category)

```python
supervised = dataset 
supervised["cluster_group"] = y_label
supervised  # new dataset 
```

## save the dataset in csv file 

```python
supervised.to_csv("hdbscan_cluster.csv",index=False)
```
