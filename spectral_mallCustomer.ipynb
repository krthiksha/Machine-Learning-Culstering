# Step 1 

## read the dataset 

```python
import pandas as pd
dataset = pd.read_csv("Mall_Customers.csv")
dataset
```

## Input 
### Select the 2 important columns (which has the greater influence over the clustering for type of customers)

```python
x = dataset.iloc[:, [3,4]].values
print(x , type(x))
```

## Standardization

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
x
```

## Model Creation 

```python
# Model creation
from sklearn.cluster import SpectralClustering
""" based on Domain knowledge, assigned : n_clusters=5 """
SpectralClustering_model = SpectralClustering(n_clusters=5,gamma=1.0, affinity='rbf',n_neighbors=10,assign_labels='kmeans',  random_state=0)
y_label = SpectralClustering_model.fit_predict(x)
y_label
```

```python
# RESULTS
affinity_matrix = SpectralClustering_model.affinity_matrix_
print("affinity_matrix: ", affinity_matrix)
```

```python
# to list the methods and attributes of any algorithm - use dir() function
# dir(SpectralClustering_model)
```

## Create the new dataset by adding  the cluster group (type of customer category)

```python
supervised = dataset 
supervised["cluster_group"] = y_label
supervised  # new dataset 
```

## save the dataset in csv file 

```python
supervised.to_csv("SpectralClustering_cluster.csv",index=False)
```

# Step 3 - Evaluation metric (clustering)

```python
from sklearn.metrics import silhouette_score
score = silhouette_score(x, y_label)
print("Silhouette Score:", score)
# -1 → 1 (higer is better (1))
```
