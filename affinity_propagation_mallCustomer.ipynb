# AFFINITY PROPAGATION clustering - Unsupervised learning
## Problem statement: Malls customer dataset (Target Marketting)

# Step 1 

## read the dataset 

```python
import pandas as pd
dataset = pd.read_csv("Mall_Customers.csv")
dataset
```

## Input 
### Select the 2 important columns (which has the greater influence over the clustering for type of customers)

```python
x = dataset.iloc[:, [3,4]].values
print(x , type(x))
```

## Standardization

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
x
```

## Model Creation 

```python
from sklearn.cluster import AffinityPropagation
"""
damping -  range [0.5, 1.0]

"""
AffinityPropagation_model = AffinityPropagation(damping=0.5,preference=None, affinity='euclidean')
y_label = AffinityPropagation_model.fit_predict(x)
y_label
```

```python
# RESULTS
exemplar = AffinityPropagation_model.cluster_centers_indices_
affinity_matrix = AffinityPropagation_model.affinity_matrix_
print("exemplar: ", exemplar)
print("affinity_matrix: ", affinity_matrix)
```

```python
# to list the methods and attributes of any algorithm - use dir() function
# dir(AffinityPropagation_model)
```

## Create the new dataset by adding  the cluster group (type of customer category)

```python
supervised = dataset 
supervised["cluster_group"] = y_label
supervised  # new dataset 
```

## save the dataset in csv file 

```python
supervised.to_csv("AffinityPropagation_cluster.csv",index=False)
```

# Step 3 - Evaluation metric (clustering)

```python
from sklearn.metrics import silhouette_score
score = silhouette_score(x, y_label)
print("Silhouette Score:", score)
# -1 → 1 (higer is better (1))
```
